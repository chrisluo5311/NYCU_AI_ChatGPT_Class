```python
from flask import Flask
from flask_ngrok import run_with_ngrok
app = Flask(__name__)
run_with_ngrok(app)

@app.route('/', methods=['GET'])
def index():
    return "這個頁面是首頁!"

@app.route('/hello', methods=['GET'])
def hello():
    return "這個頁面是歡迎頁面!"

if __name__ == '__main__':
    app.run()
```
