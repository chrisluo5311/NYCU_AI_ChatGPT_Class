```python
'''
ChatGPT真正傳出去的tokens與接收的訊息大概長這樣
 {"token": "<|im_end|>"}, "\n", {"token": "<|im_start|>"},
 "user\n以下對話請用繁體中文回答問題",
 {"token": "<|im_end|>"}, "\n", {"token": "<|im_start|>"},
 "assistant\nOpenAI是一家人工智......",
'''

#因此我們可以計算一個

def tokens_in_messages(messages):
    encoder = tiktoken.encoding_for_model("gpt-3.5-turbo")
    totals = 0
    for message in messages:
        for k in message:
            if k == "content":
                #這個是加上ChatGPT在發送問題使用的基本tokens，如<|im_start|>與<|im_end|>
                totals += 5
                totals += len(encoder.encode(message[k]))
    #這個是加上ChatGPT在回答問題使用的基本tokens，如<|im_start|>與<|im_end|>
    totals += 3
    return totals
```

```python
#可證實與上方的promt_tokens數量一樣
tokens_num = tokens_in_messages([
        {"role":"system", "content":"以下對話請用繁體中文回答問題"},
        {"role":"user","content":"請介紹OpenAI是什麼樣的公司"}
    ])
print(tokens_num)
```

```python
#因此其實問問題前就可以大致估算這個問題提出去的成本
#gpt-3.5-turbo 的tokens單價是 0.002(美金)/1000
#假設我們提出的問題長度為ask_tokens_num，限定其回答的長度為ans_tokens_num(500)，則成本如下：

ask_tokens_num = tokens_num
ans_tokens_num = 500
twd = 30.68

price = (ask_tokens_num+ans_tokens_num) / 1000 * 0.002 * twd
print(price)
```

```python
# 把以上程式固定成一個函式方便程式調用
def chatgpt_QA(Q):
    response = openai.Completion.create(
    model="text-davinci-003",
    prompt=Q,
    temperature=0,
    max_tokens=500,
    top_p=1,
    frequency_penalty=0.5,
    presence_penalty=0
    )
    # 返回答案
    return response["choices"][0]["text"].strip()
```

```python
Q = input("請輸入您的問題：")
print(chatgpt_QA(Q))
```

```python
def chatgpt_QA_loop(messages):
    response = openai.ChatCompletion.create(
            model = "gpt-3.5-turbo",
            messages = messages
        )
    ans = response["choices"][0]["message"]["content"]
    return ans
```

```python
hists=[] # 用來儲存對話紀錄的
backtrace=2 # 記錄幾組對話

def chat(sys_msg, user_msg):
    while len(hists)>=2*backtrace: #超過記錄限制
        hists.pop(0) #移除最舊的紀錄
    hists.append({"role":"user", "content":user_msg})
    reply = chatgpt_QA_loop(hists + [{"role":"system", "content":"接下來請你當作自己是一位"+sys_msg}])
    #while len(hists)>=2*backtrace: # 超過記錄限制
    #    hists.pop(0) # 移除最舊紀錄
    hists.append({"role":"assistant", "content":reply})
    return reply
```
