### 範例1，假設目標函數 f(x)=$X^2$，而非MSE，請使用梯度下降法取最小值

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 目標函數 f(x) = x^2
def func(x):
    return x ** 2
```

```python
# 目標函數的一階導數
def dfunc(x):
    return 2*x
```

```python
# 梯度下降
# x_start:起始點
# df: 目標導數的一接導數
# epoch: 回合數 
# lr: 學習率
def gradient_dscent(x_start,df,epoch,lr):
    tangent_points = np.zeros(epoch+1)
    # 先儲存第一個起始點
    tangent_points[0] = x_start
    x = x_start
    for i in range(epoch):
        dx = df(x)
        x = x - (dx*lr)
        tangent_points[i+1] = x
    return tangent_points
```

```python
# Hyperparameters
x_start = 5
epoch = 10
lr = 0.3

# 梯度下降法
tangent_points = gradient_dscent(x_start,dfunc,epoch,lr)
print(np.around(tangent_points,2))
```

```python
plt.figure(figsize=(12,8))
# 創立一個t的ndarray由-6~6，步伐是0.01
t = np.arange(-6.0, 6.0, 0.01)
# x:t, y:func(x), color=blue
plt.plot(t,func(t),c='b')
```
